```python
class Employee:
    company = "HP"
    def __init__(self,name,salary):
        
      self.name = name
      self.salary = salary
        
    def printDetails(self):
        print(f"the company of {self.name} is {self.company} with the salary of {self.salary}")

    @staticmethod
    def printTime():
        print("the time is now over")

e = Employee("Anuj" , "23000")
e.printDetails()
e.printTime()
```

```python
class Employee:
    company = "HP"
    def __init__(self,name,salary):
        
      self.name = name
      self.salary = salary
        
    def printDetails(self):
        print(f"the company of {self.name} is {self.company} with the salary of {self.salary}")

    @staticmethod
    def printTime():
        print("the time is now over")

    @classmethod
    def printClassDetails(cls):
        print(f"the company is {cls.company}")

e = Employee("Anuj" , "23000")
e.printDetails()
e.printTime()
e.printClassDetails()
```
